```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# importing the dataset

df = pd.read_csv('/content/dataset.csv')
df.head()
```

```python
# an extra 'row number' column was already present. removing that column

df = df.iloc[: , 1:]
df.head()
```

```python
# shape of the dataset

df.shape
```

```python
# descriptive statistics of the dataset

df.describe()
```

```python
# number of missing values for every column

df.isnull().sum()
```

```python
# percentage of missing values in every column

df.isnull().sum() / df.shape[0] * 100
```

```python
# total number of missing values in the entire dataset

df.isnull().sum().sum()
```

```python
percent = df.isnull().sum().sum() / df.count().sum() * 100
print(f'Total Percentage of missing values: {percent}%')
```

```python
# getting the data types of the columns

df.dtypes
```

```python
# Activity is the only categorical column
# subject is the number associated with the particular subject(person)
# all other columns are numeric(floats)
```

```python
# Subject is present in groups i.e. values for a given subject are together
# same is the case for activity, for a given subject, values for a given activity are together
# hence for missing value imputation in case of these two columns, we use the ffill function

df['Activity'].fillna(method='ffill', inplace = True)
df['subject'].fillna(method='ffill',inplace = True)
```

```python
# re-checking if there are still any missing values present

df['Activity'].isnull().sum(), df['subject'].isnull().sum()
```

```python
# defining a varialble which stores the list of the remaining columns so that we can traverse through the list
# filling the missing value for the numeric columns, by taking the mean of values in the same subject and activity group 

col = list(df)
col = col[:-2]
for i in col:
  df[i] = df.groupby(['Activity','subject'])[i].transform(lambda x: x.fillna(x.mean()))
```

```python
# all the missing values from the dataset have been eliminated

df.isnull().sum().sum()
```
